## Mini-Bactch GD
Everything in the plotting code and the sigmoid class (SN) remains the same. We are making changes to the cell which controls which paramenters to choose from the SN class.  
An explanation of how the Mini-Bactch GD works.

```python
#used for 3D Plotting
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

from matplotlib import cm
import matplotlib.colors

# For animated plots
from matplotlib import animation, rc
from IPython.display import HTML

import numpy as np
```

```python
class SN:
  
  def __init__(self, w_init, b_init, algo):
    self.w = w_init
    self.b = b_init
    self.w_h = []
    self.b_h = []
    self.e_h = []
    self.algo = algo

  def sigmoid(self, x, w=None, b=None):
    if w is None:
      w = self.w
    if b is None:
      b = self.b
    return 1. / (1. + np.exp(-(w*x + b))) # here w and b are scalar inputs

  def error(self, X, Y, w=None, b=None):
    if w is None:
      w = self.w
    if b is None:
      b = self.b
    err = 0
    for x, y in zip(X, Y):
      err += 0.5 * (self.sigmoid(x, w, b) - y) ** 2
    return err

  def grad_w(self, x, y, w=None, b=None):
    if w is None:
      w = self.w
    if b is None:
      b = self.b
    y_pred = self.sigmoid(x, w, b)
    return (y_pred - y) * y_pred * (1 - y_pred) * x

  def grad_b(self, x, y, w=None, b=None):
    if w is None:
      w = self.w
    if b is None:
      b = self.b
    y_pred = self.sigmoid(x, w, b)
    return (y_pred - y) * y_pred * (1 - y_pred)
  
  def fit(self, X, Y, 
          epochs=100, eta=0.01, gamma=0.9, mini_batch_size=100, eps=1e-8,  
          beta=0.9, beta1=0.9, beta2=0.9
         ):
    self.w_h = []
    self.b_h = []
    self.e_h = []
    self.X = X
    self.Y = Y
    
  def fit(self, X, Y, 
          epochs=100, eta=0.01, gamma=0.9, mini_batch_size=100, eps=1e-8,  
          beta=0.9, beta1=0.9, beta2=0.9
          ):
    self.w_h = []
    self.b_h = []
    self.e_h = []
    self.X = X
    self.Y = Y
    
    if self.algo == 'GD':
      # Going epochs number of times through the data after we seeing one round of seeing the data we update the paramemters
      for i in range(epochs):
        dw, db = 0, 0
        for x, y in zip(X, Y):
          # We are accumalating here but only after we have gone through the epochs we do a single update
          dw += self.grad_w(x, y)
          db += self.grad_b(x, y)
        self.w -= eta * dw / X.shape[0]
        self.b -= eta * db / X.shape[0]
        self.append_log()
    
    # Mini-Barch is best contrasted with the vanilla GD algo, 
    # at one extreme of the mini batch is the stocastic gd algo which is bacically mini batch with batch size of 1.  
    elif self.algo == 'MiniBatch':
      # We would like to do the updates more frequently.
      for i in range(epochs):
        dw, db = 0, 0
        points_seen = 0
        for x, y in zip(X, Y):
          dw += self.grad_w(x, y)
          db += self.grad_b(x, y)
          # We are keeping the track of the number of points seen
          points_seen += 1
          # If the number of points seens becomes a multiple of the number mini_batch_size (the no of times after we want to make the update), 
          # then we would go ahead and update the parameters
          # When the mini_batch_size is 1 it is called stocastic GD, if we make the mini_batch_size equla to the training data size, then it is same as vanilla GD
          if points_seen % mini_batch_size == 0:
            self.w -= eta * dw / mini_batch_size
            self.b -= eta * db / mini_batch_size
            self.append_log()
            dw, db = 0, 0
        
    elif self.algo == 'Momentum':
      v_w, v_b = 0, 0
      for i in range(epochs):
        dw, db = 0, 0
        for x, y in zip(X, Y):
          dw += self.grad_w(x, y)
          db += self.grad_b(x, y)
        v_w = gamma * v_w + eta * dw 
        v_b = gamma * v_b + eta * db
        self.w = self.w - v_w
        self.b = self.b - v_b
        self.append_log()
        
    elif self.algo == 'NAG':
      v_w, v_b = 0, 0
      for i in range(epochs):
        dw, db = 0, 0
        v_w = gamma * v_w
        v_b = gamma * v_b
        for x, y in zip(X, Y):
          dw += self.grad_w(x, y, self.w - v_w, self.b - v_b)
          db += self.grad_b(x, y, self.w - v_w, self.b - v_b)
        v_w = v_w + eta * dw
        v_b = v_b + eta * db
        self.w = self.w - v_w
        self.b = self.b - v_b
        self.append_log()

  def append_log(self):
    self.w_h.append(self.w)
    self.b_h.append(self.b)
    self.e_h.append(self.error(self.X, self.Y))
```

```python
# Changes to be made here to run different types of Algos according to the SN class.
# X = np.asarray([0.5,2.5])
# Y = np.asarray([0.2,0.9])

# w_init = -4
# b_init = 0

# Changing the data for a diffrerent error surface
X = np.asarray([3.5, 0.35, 3.2, -2.0, 1.5, -0.5])
Y = np.asarray([0.5, 0.50, 0.5, 0.5, 0.1, 0.3])

algo = 'MiniBatch'

w_init = 2.1
b_init = 4.0

w_min = -7
w_max = 5

b_min = -7
b_max = 5

epochs = 200
# For stocastic GD
# mini_batch_size = 1
mini_batch_size = 6
gamma = 0.9
eta = 5

animation_frames = 20

plot_2d = True  # contour plots
plot_3d = False 
```

__Problem with Momentum GD__  
The problem with vanilla GD is that, in regions of low gradiends it is unable to reinforce the directions it should move.

```python
sn = SN(w_init, b_init, algo)
sn.fit(X, Y, epochs=epochs, eta=eta, gamma = gamma, mini_batch_size=mini_batch_size)
plt.plot(sn.e_h, 'r')
plt.plot(sn.w_h, 'b')
plt.plot(sn.b_h, 'g')
plt.show()
```

```python
def plot_animate_3d(i):
  i = int(i*(epochs/animation_frames))
  line1.set_data(sn.w_h[:i+1], sn.b_h[:i+1])
  line1.set_3d_properties(sn.e_h[:i+1])
  line2.set_data(sn.w_h[:i+1], sn.b_h[:i+1])
  line2.set_3d_properties(np.zeros(i+1) - 1)
  title.set_text('Epoch: {: d}, Error: {:.4f}'.format(i, sn.e_h[i]))
  return line1, line2, title
```

__Observations__  
With mini batch size 1 - There are a lot of ossilations while moving towards the right direction, because each time we do update for 1 data item. The total no. of epochs to reach to the global minima might be lesser but it is highly noisy.

As we increase the batch size the noise while moving towards the minima decreases.

sometimes we have to break the data into mini batch sizes just to fit inside the limits of the gpu memory say for example a video input we man not simply be able to do a whole pass of a epoch in those cases too we can use mini batch to break the data into smaller chunks.
